```python
import math
class Point:
  def __init__(self, x,y):
    self.x = x
    self.y = y

  def distance(self,other):
    dx = self.x - other.x
    dy = self.y - other.y
    return (math.sqrt(dx*dx + dy*dy))

class Shape:
  def __init__(self,name):
    self.name = name

  def area(self):
    print("Area is: ", end="")

  def perimeter(self):
    print("Perimeter is: ", end="")

  def draw(self,x,y):
    print(x, end=",")
    print(y)

class Line(Shape):
  def __init__(self,p1,p2):
    super().__init__("Line")
    self.p1 = p1
    self.p2 = p2

  def area(self):
    print(0)

  def perimeter(self):
    print(self.p1.distance(self.p2))

  def draw(self):
    super().draw(self.p1.x,self.p1.y)
    super().draw(self.p2.x,self.p2.y)

class Circle(Shape):
  def __init__(self,p1,p2):
    super().__init__("Circle")
    self.p1 = p1
    self.p2 = p2
    self.radius = (self.p1.distance(self.p2))/2

  def area(self):
    Shape.area(self)
    print(math.pi*self.radius*self.radius)

  def perimeter(self):
    Shape.perimeter(self)
    print(2*math.pi*self.radius)

  def draw(self):
    super().draw(self.p1.x,self.p1.y)
    super().draw(self.p2.x,self.p2.y)

class Triangle(Shape):
  def __init__(self,p1,p2,p3):
      super().__init__("Triangle")
      self.p1 = p1
      self.p2 = p2
      self.p3 = p3

  def area(self):
    Shape.area(self)
    print(0.5*(self.p1.distance(self.p2)*(self.p1.distance(self.p3))))

  def perimeter(self):
    Shape.perimeter(self)
    print((self.p1.distance(self.p2) + self.p2.distance(self.p3) + self.p3.distance(self.p1)))

  def draw(self):
    super().draw(self.p1.x,self.p1.y)
    super().draw(self.p2.x,self.p2.y)
    super().draw(self.p3.x,self.p3.y)

class Rectangle(Shape):
  def __init__(self,width,height):
    super().__init__("Rectangle/Square")
    self.width = width
    self.height = height

  def perimeter(self):
    Shape.perimeter(self)
    print(2*(self.width+self.height))

  def area(self):
    Shape.area(self)
    print(self.width*self.height)

  def draw(self):
    print(f"Width = {self.width}, Height = {self.height}")

class Square(Rectangle):
  def __init__(self,width):
    super().__init__(width,width)

  def draw(self):
    print(f"Square has side = {self.width}")

class Quadilateral(Shape):
  def __init__(self,p1,p2,p3,p4):
    super().__init__("Quad")
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4

  def perimeter(self):
    Shape.perimeter(self)
    print(self.p1.distance(self.p2) + self.p2.distance(self.p3) +
      self.p3.distance(self.p4) + self.p4.distance(self.p1))

  def area(self):
    x1, y1 = self.p1.x, self.p1.y
    x2, y2 = self.p2.x, self.p2.y
    x3, y3 = self.p3.x, self.p3.y
    x4, y4 = self.p4.x, self.p4.y
    Shape.area(self)
    print(abs((x1*y2 - x2*y1) + (x2*y3 - x3*y2) +
    (x3*y4 - x4*y3) + (x4*y1 - x1*y4)) / 2)

  def draw(self):
    super().draw(self.p1.x,self.p1.y)
    super().draw(self.p2.x,self.p2.y)
    super().draw(self.p3.x,self.p3.y)
    super().draw(self.p4.x,self.p4.y)

#       Main

p1 = Point(0, 0)
p2 = Point(4, 0)
p3 = Point(2, 3)
p4 = Point(4, 3)

print("\nLINE")
line = Line(Point(0, 0), Point(3, 4))
line.draw()
line.area()
line.perimeter()

print("\nCIRCLE")
circle = Circle(Point(0, 0), Point(6, 0))
circle.draw()
circle.area()
circle.perimeter()

print("\nTRIANGLE")
triangle = Triangle(p1, p2, p3)
triangle.draw()
triangle.area()
triangle.perimeter()

print("\nRECTANGLE")
rect = Rectangle(5, 3)
rect.draw()
rect.area()
rect.perimeter()

print("\nSQUARE")
square = Square(4)
square.draw()
square.area()
square.perimeter()

print("\nQUADRILATERAL")
quad = Quadilateral(p1, p2, p4, p3)
quad.draw()
quad.area()
quad.perimeter()
```

```python
class Array:
  def __init__(self,size):
    self.size = size
    self.items = 0
    self.arr = [None]*size

  def add(self,n,index):
    if(self.items >= self.size):
      return
    self.arr[index] = n
    if(self.items < index):
      self.items = index+1

  def delete(self, index):
    if(self.items > 0):
      self.arr[index] = None
      self.items -= 1
    return

  def display(self):
    for i in range(self.items):
      print(self.arr[i])


class Stack(Array):
  def __init__(self,size):
    super().__init__(size)

  def add(self,n):
    if(self.items >= self.size):
      return
    self.arr[self.items]  = n
    self.items += 1

  def delete(self):
    if(self.items > 0):
      self.arr[self.items] = None
      self.items -= 1
    return

class Queue(Array):
  def __init__(self,size):
    super().__init__(size)
    self.front = 0

  def add(self,n):
    if(self.items >= self.size):
      return
    self.arr[self.items]  = n
    self.items += 1

  def delete(self):
    if(self.items > 0):
      self.arr[self.front] = None
      self.items -= 1
      self.front += 1
    return

  def display(self):
    for i in range(self.front,self.items+1):
      print(self.arr[i])

print("Array\n")
arr = Array(5)
arr.add(2,0)
arr.add(1,1)
arr.add(0,4)
arr.display()
print("\nAfter deletion:")
arr.delete(2)
arr.display()

print("\nStack\n")
stk = Stack(5)
stk.add(1)
stk.add(2)
stk.add(3)
stk.display()
print("\nAfter deletion:")
stk.delete()
stk.display()

print("\nQueue\n")
q = Queue(5)
q.add(1)
q.add(2)
q.add(3)
q.add(4)
q.display()
print("\nAfter deletion")
q.delete()
q.display()
```
